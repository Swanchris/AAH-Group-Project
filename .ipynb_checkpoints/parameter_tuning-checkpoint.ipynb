make new df for each k value

```python
k = {}
for i in range(1,11):
    k[i] = k_data.loc[i]
    k[i] = k[i].reset_index(drop = "true")
```

plotting (eroneous) of changes as k varies (it doesnt change)

```python
t = range(len(k[1]))
data1 = k[1]['approximation ratio']
data2 = k[1]['time taken']

fig, ax1 = plt.subplots()

color = 'tab:green'
ax1.set_xlabel('instances')
ax1.set_ylabel('approximation ratio', color=color)
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('time taken', color=color)  # we already handled the x-label with ax1
ax2.plot(t, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
